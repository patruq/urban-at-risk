# Fire Incident Reporting

## `fire_df`

```python
path = r"../data/fire-incident-reporting/"
files = glob.glob(path + "/*.csv")

data_list = []

for file in files:
    df = pd.read_csv(file, index_col=None, header=0)
    data_list.append(df)
    
fire_df = pd.concat(data_list, axis=0, ignore_index=True)
```

```python
print(fire_df.shape)
fire_df.head(10)
```
